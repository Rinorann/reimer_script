```python
# 1. Инициализируем стейт, прокидывая конфиг из виджетов app
# app — это экземпляр ReimerApp из предыдущей ячейки
state = state_manager.init_state(config=app.get_config())
history = [deepcopy(state)] 
```

```python
print("🔬 FERROZINE CLI v2.0")
print("Команды: wizard, add, cal, bl, load, fit, plan, show, undo, exit")
print("-" * 50)

while True:
    try:
        cmd = input('> ').strip()
        
        if not cmd:
            continue
            
        cmd_lower = cmd.lower()

        # --- Спец-команда: UNDO ---
        if cmd_lower == 'undo':
            if len(history) > 1:
                history.pop()
                state = deepcopy(history[-1])
                print("✅ Отменено (вернулись на шаг назад)")
            else:
                print("⚠️ Нечего отменять")
            continue
            
        # --- Спец-команда: EXIT ---
        if cmd_lower == 'exit':
            print("👋 Лаборатория закрыта. Результаты сохранены в переменной 'state'.")
            break
            
        # --- Спец-команда: WIZARD (Режим долбоёба) ---
        if cmd_lower == 'wizard':
            # Визард сам управляет своим состоянием внутри
            state, report = executors.execute_wizard(state)
            history.append(deepcopy(state))
            print(report)
            continue

        # --- ОБЫЧНЫЙ РЕЖИМ (Парсинг -> Экзекьютор) ---
        command = parsers.parse_command(cmd)
        
        # Выполняем команду
        new_state, report = executors.execute_command(state, command)
        
        # Обновляем историю и стейт, если это не просто просмотр (show)
        action = command.get('action', '')
        if action != 'show':
            state = new_state
            history.append(deepcopy(state))
        
        # Выводим результат (таблицу, рецепт или расчеты)
        print(report)

    except Exception as e:
        print(f"❌ ОШИБКА: {str(e)}")
        print("Попробуйте еще раз или введите 'undo', если что-то сломалось.")
```
