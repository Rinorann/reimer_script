```python
import pandas as pd
from copy import deepcopy
import numpy as np
import parsers
import utils
import executors 
```

REPL

```python
state = utils.initial_state()
history = [deepcopy(state)]
```

```python
# @title 🎮 Запуск эксперимента { display-mode: "form" }

print("🔬 96-луночный планшет")
print("Команды: add, edit, erase, rename, delete, show, undo")
print('Пример: add "образец 1" dil 20 B1:B3, dil 10 C1:C3')
print("-" * 40)

while True:
    cmd = input('> ')
    
    if cmd.lower() == 'undo':
        if len(history) > 1:
            history.pop()
            state = deepcopy(history[-1])
            print("✅ Отменено")
        else:
            print("⚠️ Нечего отменять")
        continue
        
    if cmd.lower() == 'exit':
        print("👋 Пока")
        break
        
    if not cmd.strip():
        continue
    
    # Парсим и выполняем
    command = parsers.parse_command(cmd)
    state, report = executors.execute_command(state, command)
    
    # Сохраняем в историю
    history.append(deepcopy(state))
    
    # Выводим отчет
    print(report)
    print()
```
